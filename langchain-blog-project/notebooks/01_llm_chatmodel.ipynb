# 01 — LLMs and Chat Models
This notebook demonstrates how to use OpenAI's Chat Model through LangChain's unified interface.

## Setup

```
# Install dependencies
# !pip install langchain langchain-openai langchain-core python-dotenv
```

```
import os
from dotenv import load_dotenv
load_dotenv()  # Loads OPENAI_API_KEY from .env file
```

## 1. Basic Chat Model Call

```
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage

# Initialise the model
# temperature=0 → deterministic output (best for factual tasks)
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

messages = [
    SystemMessage(content="You are a helpful Python tutor."),
    HumanMessage(content="Explain list comprehension in one sentence.")
]

response = llm.invoke(messages)
print(response.content)
```

## 2. Viewing Token Usage

```
# LangChain surfaces token usage metadata
response_with_meta = llm.invoke(messages)
print("Content     :", response_with_meta.content)
print("Model       :", response_with_meta.response_metadata.get("model_name"))
print("Token usage :", response_with_meta.response_metadata.get("token_usage"))
```

## 3. Streaming Output

```
# Stream tokens as they are generated
for chunk in llm.stream([HumanMessage(content="Count from 1 to 5 slowly.")]):
    print(chunk.content, end="", flush=True)
print()  # newline after stream ends
```

## 4. Batch Calls

```
# Send multiple prompts in one batch call (more efficient than looping)
questions = [
    [HumanMessage(content="What is a decorator in Python?")],
    [HumanMessage(content="What is a generator in Python?")],
    [HumanMessage(content="What is a context manager in Python?")],
]

results = llm.batch(questions)
for i, r in enumerate(results, 1):
    print(f"Q{i}: {r.content[:100]}\n")
```

---
**Key takeaway:** LangChain's `ChatOpenAI` wraps the raw API and exposes `.invoke()`, `.stream()`, and `.batch()` — the same interface works for any provider.
